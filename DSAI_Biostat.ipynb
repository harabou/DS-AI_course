#6. 分散分析（ANOVA）

```python
fpath = 'https://raw.githubusercontent.com/harabou/DS-AI_course/main/%2301/hosp.csv'
#１行目（pythonでは0行と指定）を見出しにする。日本語を含むためエラーが出ることが多いので、[Shift-jis]で読み込み
hosp = pd.read_csv(fpath,header=0, encoding='shift-jis')
```

```python
hosp.head()
```

```python
from statsmodels.formula.api import ols
model = ols('Time ~ Hosp', data=hosp).fit()
model.summary()
```

```python
import statsmodels.api as sm

#引数typは、平方和のタイプ指定（いわゆるANOVA Type I, II, IIIなどと分類されるもの）
#今回はtyp=2としたが、今回のように全ての溶媒に対して同数（5サンプル）のデータがある場合は
#結果に差異は生じない
anova = sm.stats.anova_lm(model, typ=3)
anova
```

```python
import numpy as np
import pandas as pd
from scipy import stats

hospA = np.array([500, 420, 650, 550, 480])
hospB = np.array([400, 420, 460, 540, 390])
hospC = np.array([350, 400, 320, 280, 490])

#後々のために、一度データフレームにまとめておく
data = pd.DataFrame({
    'A' : hospA,
    'B' : hospB,
    'C' : hospC
})

fvalue, pvalue = stats.f_oneway(data['A'], data['B'], data['C'])
# *以下のようにNumPy配列を直接与えることも可能
#  fvalue, pvalue = stats.f_oneway(solventA, solventB, solventC)

print(f"F-val:{fvalue:.4},  P-val:{pvalue:.4}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib
from urllib import request
import lifelines
from lifelines import KaplanMeierFitter
```

```python
import pandas as pd
fpath = 'https://raw.githubusercontent.com/harabou/DS-AI_course/main/%2301/gehan.csv'
#１行目（pythonでは0行と指定）を見出しにする。日本語を含むためエラーが出ることが多いので、[Shift-jis]で読み込み
gehan = pd.read_csv(fpath,header=0, encoding='shift-jis')
```

```python
gehan.head()
```

```python
# Kaplan-Meier-Fitter
kmf = KaplanMeierFitter()
kmf.fit(durations = gehan.week, event_observed = gehan.remiss)
```

```python
kmf.event_table
```

```python
# Calculate survival probability for given time
print("Survival probablity for t=0: ", kmf.predict(0))
print("Survival probablity for t=5: ", kmf.predict(5))
print("Survival probablity for t=11: ", kmf.predict(11))
print("Survival probablity for t=840: ", kmf.predict(840))
```

```python
# Get survial full list
print(kmf.survival_function_)
```

```python
# Plot the survival graph
kmf.plot()
plt.title("Kaplan-Meier")
plt.xlabel("Number o days")
plt.ylabel("Probablity of survival")
plt.show()
```

```python
# Get survial time median
survival_time_median = kmf.median_survival_time_
print(" The median survial time: ", survival_time_median)
```

```python
# Plot survival function with confidence interval
survival_confidence_interval = kmf.confidence_interval_survival_function_
plt.plot(survival_confidence_interval['KM_estimate_lower_0.95'], label="Lower")
plt.plot(survival_confidence_interval['KM_estimate_upper_0.95'], label="Upper")
plt.title("Survival Function with confidence interval")
plt.xlabel("Number of days")
plt.ylabel("Survial Probablity")
plt.legend()
plt.show()
```

```python
# Cumulative survival
cumu_surv = kmf.cumulative_density_
# Plot cumulative density
kmf.plot_cumulative_density()
plt.title("Cumulative density")
plt.xlabel("Number of days")
plt.ylabel("Probablity of death")
plt.show()
```

```python
# KM for Male and Female
kmf_1 = KaplanMeierFitter()
kmf_0 = KaplanMeierFitter()
d1 = gehan.query("drug == 1")
d0 = gehan.query("drug == 0")
kmf_1.fit(durations=d1.week, event_observed=d1.remiss, label="d1")
kmf_0.fit(durations=d0.week, event_observed=d0.remiss, label="d0")
```

```python
# Plot survival function
kmf_1.plot()
kmf_0.plot()
plt.title("KM Survival Probaility: Active vs placebo")
plt.xlabel("Week")
plt.ylabel("Remiss Probability")
plt.show()
```

```python
from lifelines.statistics import logrank_test

time_d1 = d1.week
event_d1 = d1.remiss

time_d0 = d0.week
event_d0 = d0.remiss

results = logrank_test(time_d1, time_d0, event_d1, event_d0)
results.print_summary()
print("P-value: ", results.p_value)
```
